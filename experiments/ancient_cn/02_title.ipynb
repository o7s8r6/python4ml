````python
# I'm sure the following ns isn't what we want
# pardon my variable name, they are just useless in our task

useless_ns = [4,828,10,8,106,12,2600,6,114]
````

Remove useless files

````python
import os
for ns in useless_ns:
    subdf = titles[titles.ns==ns]
    for uuid in subdf.uuid:
        os.system(f"rm -rf /Users/salvor/cn_text/{uuid}.txt")
    print(f"ns {ns} files removed")
````

````python
titles = titles[titles.ns.isin(set(titles.ns.unique()) - set(useless_ns))]
````
