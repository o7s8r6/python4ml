```python
def slicing(x):
    return x[:x.find("}}")+2],x[x.find("}}")+2:],[x[:x.find("}}")+2],]

def find_emb(x):
    start_ = x.find("{{")
    if start_>0:
        x2 = x[start_:]
        if x2[2:].find("{{")<0:
            # has not sub emb
            prev_str,res_str,res_list = slicing(x2)
            next_str, next_list = find_emb(res_str)
            return next_str,res_list+next_list
        else:
            # has other embs
            start2 = x2[2:].find("{{")
            end1 = x2[2:].find("}}")
            if start2> end1:
                # another emb
                prev_str,res_str,res_list = slicing(x2)
                next_str,next_list = find_emb(x2[2:])
                return next_str,res_list+next_list
            else:
                # sub emb
                residual,emb_list = find_emb(x2[2:])
                prev_str,res_str,res_list = slicing(x2[2:])
                return prev_str+res_str,emb_list+res_list
    return x,[]
```

```python
def clean_emb(x):
    if (x[:2]=="{{") and (x[-2:]=="}}"):
        x = x[2:-2]
    divids = x.split("|")
    k = divids[0]
    v = divids[1:]
    if len(v)==0:
        return {k:None}
    if len(v)==1:
        return {k:v}
    else:
        if "=" in v[0]:
            return {k:dict((i.split("=")[0],i.split("=")[1]) for i in v)}
        else:
            return {k:list(v)}
```
