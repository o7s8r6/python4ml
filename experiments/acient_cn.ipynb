```
from pathlib import Path
DATA = Path(".")
text = DATA/"zhwikisource-20200301-pages-articles-multistream.xml"
```

```
class XMLGen:
    def __init__(self,path):
        self.f = open(path,mode="r")

    def __iter__(self):
        rt = ""
        while True:
            line = self.f.readline()
            if line:
                if "<page>" in line:
                    rt = ""
                rt+=line
                if "</page>" in line:
                    yield rt
            else:
                break
```

```
xg = XMLGen(text)
```

```
i = 0
for xml in xg:
    print(xml)
    i+=1
    if i>10:break
```
